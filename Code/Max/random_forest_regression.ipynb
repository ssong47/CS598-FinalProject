# Support Vector Regression (SVR)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_x = pd.read_csv('leo_test1_10_27_2020_fdss_final_data.txt')
dataset_y = pd.read_csv('leo_test1_10_27_2020_qtm_final_data.txt')
X = dataset_x.iloc[:, :-1].values
y = dataset_y.iloc[:, 0].values
```

```python
print(X)
```

```python
print(y)
```

```python
y = y.reshape(len(y),1)
```

```python
print(y.shape)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
print(X.shape)
```

```python
print(y)
```

```python
# Train the SVR model on the whole dataset
from sklearn.svm import SVR
y = y. ravel()
regressor = SVR(kernel = 'rbf', C = 10)
regressor.fit(X, y)
```

```python
# predicting results
y_predict = sc_y.inverse_transform(regressor.predict(X))
```

```python
print(y_predict)
```

```python
# visualize results
y_original = dataset_y.iloc[:, 0].values 
plt.plot(y_original,'-')
plt.plot(y_predict,'--')

plt.title('SVR')
plt.ylabel('Angle (deg)')
plt.xlabel('Time')
plt.show()
```

```python
dataset_x4 = pd.read_csv('leo_test4_10_27_2020_fdss_final_data.txt')
dataset_y4 = pd.read_csv('leo_test4_10_27_2020_qtm_final_data.txt')
dataset_x5 = pd.read_csv('leo_test5_10_27_2020_fdss_final_data.txt')
dataset_y5 = pd.read_csv('leo_test5_10_27_2020_qtm_final_data.txt')
```

```python
full_dataset_x = pd.concat([dataset_x1, dataset_x2, dataset_x3, dataset_x4, dataset_x5])
full_dataset_y = pd.concat([dataset_y1, dataset_y2, dataset_y3, dataset_y4, dataset_y5])
full_dataset_x = full_dataset_x.iloc[:, :-1].values
full_dataset_y = full_dataset_y.iloc[:, 0].values
```

```python
full_dataset_y = full_dataset_y.reshape(len(full_dataset_y),1)
```

```python
# split training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(full_dataset_x, full_dataset_y, test_size = 0.1, random_state = 0)
```

```python
# feature scaling
fc = StandardScaler()
sc_y = StandardScaler()

X_train = fc.fit_transform(X_train)
X_test = fc.transform(X_test)

y_train = sc_y.fit_transform(y_train)
y_train = y_train.ravel()

regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)

y_pred = sc_y.inverse_transform(regressor.predict(X_test))
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

```python
#y_test = full_dataset_y.iloc[:, 0].values 
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (15,10)
plt.plot(y_test,'-')
plt.plot(y_pred,'--')

plt.title('Random Forest Regression')
plt.ylabel('Angle (deg)')
plt.xlabel('Time')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.title('Random Forest Regression')
plt.plot(y_test,'-')
plt.ylabel('Angle (deg)')
plt.xlabel('Time')

plt.subplot(2,1,2)
plt.plot(y_pred,'--')

plt.ylabel('Angle (deg)')
plt.xlabel('Time')
plt.show()
```

```python
test_error=(y_pred-y_test.reshape(-1))
print('Average error is {:4.2f} degrees'.format(np.sum(test_error)/test_error.shape[0]))

rmse=np.sqrt(test_error**2)
print('RMSE is {:4.2f} degrees'.format(np.sum(rmse)/test_error.shape[0]))
plt.plot(rmse)
plt.title('Squared Error of Test Set')
plt.xlabel('Sample')
plt.ylabel('Error (degrees)')
plt.show()
```

###### It appears to have better preformance when train the model with the whole datasets than train it with one data set. Random forest regression also has lower average error when compared to SVR.
